```python
#importing
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#importing dataset
breast_cancer=sklearn.datasets.load_breast_cancer()
```

```python
#converting it to data frame
df_breast_cancer=pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
```

```python
df_breast_cancer['result']=breast_cancer.target
df_breast_cancer.shape
```

##1**<-- Benign Tumour**
##0**<-- Malignant Tumour**

```python
df_breast_cancer.groupby('result').mean()
```

```python
#separating the features and target columns
a=df_breast_cancer.drop(columns='result',axis=1)
b=df_breast_cancer['result']
```

```python
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.1,random_state=2,stratify=b)
print(a.shape,a_train.shape,a_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(a_train,b_train)
```

```python
a_prediction=model.predict(a_train)
accuracy=accuracy_score(b_train, a_prediction)
print(accuracy)
```

```python
a_test_prediction=model.predict(a_test)
accuracy_test=accuracy_score(b_test,a_test_prediction)
print(accuracy_test)
```
